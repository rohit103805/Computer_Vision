```python
import cv2
import keras
from keras.preprocessing import image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

# Loading the Model

```python
model=keras.models.load_model(r"D:\ML_Projects\Cat-Dog\Cat_Dog_Model.h5")
```

# Making Predictions

## 1

```python
img=cv2.imread(r"D:\ML_Projects\Cat-Dog\Cat-Dog\test\cat\cat.9504.jpg")
plt.imshow(img)
plt.show()
```

```python
image_1=Image.open(r"D:\ML_Projects\Cat-Dog\Cat-Dog\test\cat\cat.9504.jpg")
print(image_1.size)
new_image=image_1.resize((64, 64))
print(new_image.size)
```

```python
Pred_image=image.img_to_array(new_image)
Pred_image=np.expand_dims(Pred_image, axis=0)
result=model.predict(Pred_image)
result=result.reshape(-1)
max=result[0]
pred=0
for i in range(0, len(result)):
    if(result[i]>max):
        pred=i
        max=result[i]
print(pred)
```
