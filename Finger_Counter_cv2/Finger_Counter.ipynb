```python
import cv2
import mediapipe as mp
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils
Imp_ids = [8, 12, 16, 20]
```

```python
camera = cv2.VideoCapture(0)
iterator = 0
while True :
    check, image = camera.read()
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)
    Count = []
    if results.multi_hand_landmarks:
        for hand in results.multi_hand_landmarks:
            Y_Points = []
            for id, lms in enumerate(hand.landmark):
                h, w, c = image.shape
                cx, cy = int(lms.x*w), int(lms.y*h)
                Y_Points.append(cy)
                if (id==4):
                    if(Y_Points[4]>Y_Points[3]):
                        Count.append(0)
                    else:
                        Count.append(1)
                if id in Imp_ids:
                    if(Y_Points[id]>Y_Points[id-2]):
                        Count.append(0)
                    else:
                        Count.append(1)
            mpDraw.draw_landmarks(image, hand, mpHands.HAND_CONNECTIONS)
    if (sum(Count)>0) :
        print(Count)

    cv2.putText(image, f'Fingers : {int(sum(Count))}', (0, 50), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 0), cv2.LINE_4)

    cv2.imshow("Hands", image)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
camera.release()
cv2.destroyAllWindows()
```
