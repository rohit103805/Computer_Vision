```python
import cv2
import numpy as np
from copy import deepcopy
from keras.models import load_model
import matplotlib.pyplot as plt
```

```python
model = load_model(r'D:\ML_Projects_PyTorch\Road_Segmentation\code_files\Road_Segmentation.h5')
```

```python
Listy = []
img = (cv2.imread(r'D:\ML_Projects_PyTorch\Road_Segmentation\data_road_224\training\image\image_224 (130).png')[:, :, ::-1])/255
img = cv2.resize(img, (224, 224))
org_img = deepcopy(img)
Listy.append(img)

Listy = np.array(Listy)

pred = model.predict(Listy)
pred = np.squeeze(pred)

red_lo=np.array([255,30,30])
red_hi=np.array([220,0,0])

hsv_1=cv2.cvtColor(pred,cv2.COLOR_BGR2HSV)
mask_1=cv2.inRange(hsv_1, red_hi, red_lo)
img[mask_1>0] = (255, 255, 255)

plt.subplot(121)
plt.imshow(org_img)
plt.subplot(122)
plt.imshow(img)
plt.show()
```
